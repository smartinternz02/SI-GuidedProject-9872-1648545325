```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

### Line graph

```python
plt.plot(df['YearsExperience'], df['Salary'])
plt.title('Predicting Salary')
plt.xlabel('Years Experience')
plt.ylabel('Salary')           
plt.show()
```

### Scatter plot

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.title('Predicting Salary')
plt.xlabel('Years Experience')
plt.ylabel('Salary')     
plt.show()
```

### Bar graph

```python
plt.bar(df['YearsExperience'], df['Salary'])
plt.title('Predicting Salary')
plt.xlabel('Years Experience')
plt.ylabel('Salary')           
plt.show()
```

```python
plt.hist(df['YearsExperience'])
plt.hist(df['Salary'])
```

### CountPlot

```python
sns.countplot(x=df["YearsExperience"], data = df)
```

```python
sns.countplot(x=df["Salary"], data = df)
```

### Heatmap

```python
sns.heatmap(df)
```

### Boxplot

```python
sns.boxplot(x='YearsExperience', y='Salary', data = df)
```

### Pairplot

```python
sns.pairplot(df)
```

### Jointplot

```python
sns.jointplot(data=df, x='YearsExperience', y='Salary')
```

### KDE

```python
sns.kdeplot(data=df, x='YearsExperience', y='Salary')
```
